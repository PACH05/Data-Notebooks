Skewness

```python
from scipy.stats import skew,kurtosis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_pos = np.random.exponential(scale=2.0, size=1000)
data_neg = -np.random.exponential(scale=2.0, size=1000)
positive_skewness = skew(data_pos)
negative_skewness = skew(data_neg)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data_pos, kde=True, bins=30, color='blue', stat='density')
plt.title(f'Distribution with Skewness: {data_skewness:.2f}')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data_neg, kde=True, bins=30, color='red', stat='density')
plt.title(f'Distribution with Skewness: {data_skewness:.2f}')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
data_ns = np.random.normal(loc=0, scale=1, size=1000)
```

```python
sns.histplot(data_ns, kde=True, bins=30, color='green', stat='density')
plt.title(f'Distribution with Skewness: {data_skewness:.2f}')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Kurtosis

```python
np.random.seed(0)
```

```python
data_meso = np.random.normal(0, 1, 1000)

data_lepto = np.random.standard_t(df=3, size=1000)

data_platy = np.random.uniform(low=-2, high=2, size=1000)
```

```python
kurt_meso = kurtosis(data_meso)
kurt_lepto = kurtosis(data_lepto)
kurt_platy = kurtosis(data_platy)
```

```python
kurtosis_df = pd.DataFrame({
    'Distribution': ['Mesokurtic (Normal)', 'Leptokurtic (t-dist)', 'Platykurtic (Uniform)'],
    'Excess Kurtosis': [kurt_meso, kurt_lepto, kurt_platy]
})
```

```python
kurtosis_df
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data_meso, kde=True, bins=30, color='blue', stat='density')

plt.title("Mesokurtic")
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.histplot(data_lepto, kde=True, bins=30, color='red', stat='density')

# Annotate skewness value
plt.title("Leptokurtic")
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data_platy, kde=True, bins=30, color='blue', stat='density')

plt.title("Platykurtic")
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.tight_layout()
plt.show()
```
